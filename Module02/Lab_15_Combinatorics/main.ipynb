```python
#Challenge 1. Write a Python program to print all permutations of the string "Ironhack"

import math

def permute(my_str):
    n = len(my_str)
    if n == 0:
        return ['']
    prev_list = permute(my_str[1:n])
    next_list = []
    for i in range(0,len(prev_list)):
        for j in range(0,n):
            new_str = prev_list[i][0:j]+my_str[0]+prev_list[i][j:n-1]
            if new_str not in next_list:
                next_list.append(new_str)
    return next_list


my_str = 'Ironhack'
l_permutations = permute(my_str)
print('Number of permutations found:', len(l_permutations))
print('All permutations found?', len(l_permutations) == math.factorial(len(my_str)))
print(l_permutations)
```

```python
from itertools import permutations

my_str = 'Ironhack'

my_str = 'Ironhack'
l_permutations = [''.join(p) for p in permutations(my_str)]
print('Number of permutations found:', len(l_permutations))
print('All permutations found?', len(l_permutations) == math.factorial(len(my_str)))
print(l_permutations)
```

```python
# Challenge 2: How many ways can you choose 3 people from a group of 17 for making the group projects teams?
 
import math

def nb_combinations(n, m):
    return math.factorial(n) / (math.factorial(m) * math.factorial(n-m))

group_size = 17
nb_people = 3
print(f'We have {nb_combinations(group_size, nb_people): .0f} ways to choose {nb_people} people from a group of {group_size}.')
```

```python
# Challenge 3: Let's imagine you need to organize the business processes in the coffee shop.
# You know that each cup can be made from 10 possible ingredients.
# But it is not profitable to use more than 5 portions of ingredients at once.
# These ingredients should not be unique (for example, you can put milk twice, and a chocolate once).
# How many different variations of coffee you can include to the menu?

import math

def nb_combinations(n, m):
    return math.factorial(m + n -1) / (math.factorial(m) * math.factorial(n-1))

def cum_combinations(n, m):
    cum = 0
    for i in range(1, m+1):
        cum += nb_combinations(n, i)
    return cum

nb_ingredients = 10
max_portions = 5

print(f'We can include {cum_combinations(nb_ingredients, max_portions): .0f} variations of coffee in the menu.')
```
