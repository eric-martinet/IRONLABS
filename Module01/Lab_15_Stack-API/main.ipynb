# StackAPI

#### Import the necessary libraries here:

```python
from stackapi import StackAPI
import datetime
import pandas as pd
```

#### Question 1: Find the questions and answers of last month.

```python
SITE = StackAPI('stackoverflow')

start_date = datetime.datetime(2022,2,1)
end_date = datetime.datetime(2022,2,28)

# Below requests will return 500 results max
questions = SITE.fetch('questions', fromdate = start_date, todate = end_date)
answers = SITE.fetch('answers', fromdate = start_date, todate = end_date)
```

```python
# Let's look at the results in a readable way
pd.DataFrame(questions["items"])
```

```python
pd.DataFrame(answers["items"])
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
questions_python_today = SITE.fetch('questions', date = datetime.date.today(), min = 5, tagged = 'python', sort = 'votes')
```

```python
most_voted_question_python_today = questions_python_today['items'][0]
most_voted_question_python_today
```
