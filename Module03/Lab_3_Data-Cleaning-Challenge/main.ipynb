```python
# LIBRAIRIES IMPORT
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# DATA IMPORT
iris = pd.read_csv('iris-data.csv')
iris.head()
```

```python
# SHAPE
iris.shape
# 150 rows, 5 columns
```

```python
# DATA TYPES
iris.dtypes
# Data types seem OK
```

```python
# STATISTICAL DESCRIPTION
iris.describe()
# Range of each variable seems OK, no obvious outlier
```

```python
# VALUE COUNTS CATEGORICAL
iris['class'].value_counts()
```

```python
# Two spelling mistakes
iris.loc[iris['class'] == 'versicolor', 'class'] = 'Iris-versicolor'
iris.loc[iris['class'] == 'Iris-setossa', 'class'] = 'Iris-setosa'
iris['class'].value_counts()
```

```python
# MISSING VALUES
iris.isna().sum()
# Only 5 missing values on petal_width_cm
```

```python
# ZOOM ON MISSING VALUES
iris[iris.petal_width_cm.isna()]
# All from class Iris-setosa
```

```python
# HOW TO FILL MISSING VALUES?
# Zoom on class Iris-setosa / distribution of petal_width_cm
print(iris[iris['class'] == 'Iris-setosa'].describe())
print(iris.loc[iris['class'] == 'Iris-setosa','petal_width_cm'].value_counts())
sns.boxplot(data = iris[iris['class'] == 'Iris-setosa'], x = 'class', y = 'petal_width_cm')
# Half of Iris-setosa items have a petal_width_cm of 0.2
```

```python
# Le'ts check correction to decide if we want to use interpolate or mode
iris.loc[iris['class'] == 'Iris-setosa'].corr()['petal_width_cm']
# Low correlation with other variables, better use mode?
```

```python
from statistics import mode
mode_iris_setosa_petal_width_cm = mode(iris.loc[iris['class'] == 'Iris-setosa','petal_width_cm'])
iris['class'].fillna(mode_iris_setosa_petal_width_cm, inplace=True)
```

```python
# CHECK DUPLICATES
iris.duplicated().value_counts()
# Two duplicates
```

```python
# Display lines with duplicates
iris.loc[iris.duplicated(keep = False) == True]
```

```python
# DROP DUPLICATES
iris.drop_duplicates(inplace=True)
```
